## Dati radar

```python
import math
def circlePoints(x, r, cx, cy):
    return math.sqrt(math.pow(r,2) - math.pow(x-cx, 2)) + cy
```

```python
agenti_format = [
          {'linea':[[0, 0], [6*13, circlePoints(6*13, 10*13, 0, 0)]],
          'agente': 'CO',
          'mediaCentraline': 50},
          {'linea':[[0, 0], [9*13, circlePoints(9*13, 10*13, 0, 0)]],
          'agente': 'NOX',
          'mediaCentraline': 100},
          {'linea':[[0, 0], [9*13, - circlePoints(9*13, 10*13, 0, 0)]],
          'agente': 'NO2',
          'mediaCentraline': 100},
         {'linea':[[0, 0], [6*13, - circlePoints(6*13, 10*13, 0, 0)]],
          'agente': 'BENZENE' ,
          'mediaCentraline': 20},
          {'linea':[[0, 0], [0*13, - circlePoints(0*13, 10*13, 0, 0)]],
          'agente': 'SO2',
          'mediaCentraline': 300},
          {'linea':[[0, 0], [-5*13, - circlePoints(-5*13, 10*13, 0, 0)]],
          'agente': 'O3',
          'mediaCentraline': 33},
          {'linea':[[0, 0], [-9*13, - circlePoints(-9*13, 10*13, 0, 0)]],
          'agente': 'PM2.5',
          'mediaCentraline': 23.20},
          {'linea':[[0, 0], [-9*13, circlePoints(-9*13, 10*13, 0, 0)]],
          'agente': 'PM10',
          'mediaCentraline': 89},
         {'linea':[[0, 0], [-5*13, circlePoints(-5*13, 10*13, 0, 0)]],
          'agente': 'NO',
          'mediaCentraline': 40}]
```

```python
df_agente_giorno = df_selected.groupby(['Chimico', pd.to_datetime(df_selected['Date']).dt.date]).sum()
```

```python
agenti = df_selected.Chimico.unique()
```

```python
dizionario_centraline = {}
```

```python
import copy
for centr in lista_centraline:
    dizionario_centraline[centr] = copy.deepcopy(agenti_format)
    print(centr)
    
    for idx, info_agente in enumerate(dizionario_centraline[centr]):
        chimico = dizionario_centraline[centr][idx]['agente']
        print (chimico)
        print (df_agente_giorno[centr][chimico].mean())
        dizionario_centraline[centr][idx]['mediaCentraline'] = df_agente_giorno[centr][chimico].mean()
```

```python
tutte_agenti = df_selected.groupby(['Chimico', pd.to_datetime(df_selected['Date']).dt.date]).sum()[lista_centraline].mean(axis=1)
```

```python
dizionario_centraline['tutte'] = copy.deepcopy(agenti_format)
```

```python
for valore in dizionario_centraline['tutte']:
        chimico = valore['agente']
        valore['mediaCentraline'] = tutte_agenti[chimico].mean()
```

```python
with open('default_radar_2018.js', 'w') as f:
    f.write('var agentiChimici = ')
    f.write(str(dizionario_centraline) + '\n')
    
```
